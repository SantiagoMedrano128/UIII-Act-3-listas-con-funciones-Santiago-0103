```python
# Sistema de Gestión Ganadero

def leer_datos():
    """Función auxiliar para capturar datos básicos."""
    nombre = input("Ingrese el nombre del animal: ")
    tipo = input("Ingrese el tipo (Vacuno, Porcino, etc.): ")
    peso = input("Ingrese el peso (kg): ")
    return {"nombre": nombre, "tipo": tipo, "peso": peso}

def agregar(lista):
    print("\n--- Agregar Nuevo Animal ---")
    animal = leer_datos()
    # Generamos un ID simple basado en el tamaño de la lista
    animal["id"] = len(lista) + 1
    lista.append(animal)
    print(f"✅ ¡{animal['nombre']} agregado con éxito!")

def listar(lista):
    print("\n--- Inventario Ganadero ---")
    if not lista:
        print("La lista está vacía.")
    else:
        for animal in lista:
            print(f"ID: {animal['id']} | Nombre: {animal['nombre']} | Tipo: {animal['tipo']} | Peso: {animal['peso']}kg")

def buscar(lista):
    print("\n--- Buscar Animal ---")
    nombre_buscar = input("Ingrese el nombre del animal a buscar: ")
    encontrado = False
    for animal in lista:
        if animal["nombre"].lower() == nombre_buscar.lower():
            print(f"🔍 Encontrado: ID {animal['id']}, Tipo: {animal['tipo']}, Peso: {animal['peso']}kg")
            encontrado = True
            break
    if not encontrado:
        print("❌ Animal no encontrado.")

def modificar(lista):
    print("\n--- Modificar Animal ---")
    id_modificar = int(input("Ingrese el ID del animal a modificar: "))
    for animal in lista:
        if animal["id"] == id_modificar:
            print("Nuevos datos:")
            nuevos_datos = leer_datos()
            animal.update(nuevos_datos)
            print("✨ Datos actualizados correctamente.")
            return
    print("❌ No se encontró un animal con ese ID.")

def borrar(lista):
    print("\n--- Borrar Animal ---")
    id_borrar = int(input("Ingrese el ID del animal a eliminar: "))
    for i in range(len(lista)):
        if lista[i]["id"] == id_borrar:
            eliminado = lista.pop(i)
            print(f"🗑️ Se ha eliminado a: {eliminado['nombre']}")
            return
    print("❌ No se encontró el ID.")

def menu():
    ganado = []  # Nuestra lista principal

    while True:
        print("\n===== SOFTWARE GANADERO SOCIO =====")
        print("1. Agregar animal")
        print("2. Listar inventario")
        print("3. Buscar animal por nombre")
        print("4. Modificar animal por ID")
        print("5. Borrar animal por ID")
        print("6. Salir")

        opcion = input("\nSeleccione una opción (1-6): ")

        if opcion == "1":
            agregar(ganado)
        elif opcion == "2":
            listar(ganado)
        elif opcion == "3":
            buscar(ganado)
        elif opcion == "4":
            modificar(ganado)
        elif opcion == "5":
            borrar(ganado)
        elif opcion == "6":
            print("Saliendo del sistema... ¡Hasta pronto, vaquero!")
            print(" programa realiaado por Santiago Medrano NL = 37")
            break
        else:
            print("⚠️ Opción no válida, intente de nuevo.")

# Ejecutar el programa
if __name__ == "__main__":
    menu()
```
